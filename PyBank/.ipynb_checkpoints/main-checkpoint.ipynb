```python
## This is a script to calculate average, maximum and minimum of monthly changes in profits/loss from bugdet_data.csv.

# Import libraries required to run the script. csv for manipulating csv data. Path for linking to directories.
import csv
from pathlib import Path

## Initialize variables that will be used in the script body.
# skip variable is used in an if condition to skip the first row when calculating monthly differences.
skip = 0

# count variable is used to find the total number of monthly changes used in the average calculation.
count = 0

# prev_row variable is used to calculate the monthly difference as the for loop iterates through the rows of the csv.
prev_row = 0

# difference variable is the value from the current iteration minus value from the previous iteration.
difference = 0

# Create an empty list to store the calculated monthly differences in for every iteration of the for loop below.
diff_list = []

# These variables are used for the calculations
average = 0
Max = 0
Min = 0
Total = 0

# Create a list to store the first column of the csv sheet to be used for identifying the month corresponding to the highest and lowest profits using the index function.
data_cols = []

# Specifiy the path to which the csv data is found.
data = Path('../../Resources/Instructions/PyBank/Resources/budget_data.csv')

# Open the csv file in reading mode as xfile or any name you prefer.
with open(data, 'r') as xfile:
    # Read the csv data with identify comma as the delimiter of the columns.
    datafiles = csv.reader(xfile, delimiter=',')
    # Skip the first row which contains header names irrelevant to the calculations to be performed below.
    header = next(datafiles)
    
    # Initite a for loop to parse through the csv data line by line.
    for row in datafiles:
        # Add 1 to the skip variable with every iteration. This will be used for if conditions.
        skip +=1
        # Calculate the cummulative sum of all the profit/losses (2nd column) as the for loop iterates through the rows.
        Total += int(row[1])
        # Store the month data from the 1st column in the data_cols list.
        data_cols.append(row[0])
        
        # If condition used here to seperate the first row from the rest when performing the calculations and to store the first prev_row value as an integer.
        if skip ==1:
            prev_row = int(row[1])
        # after the first iteration, this if condition will start calculating the monthly differences with each iteration and append it to the diff_list. 
        # Before moving to the next iteration, and only after calculating the difference between current and previous row, make current row = prev_row.    
        if skip >1: 
            count +=1            
            difference = int(row[1]) - prev_row
            diff_list.append(difference)
            prev_row = int(row[1])
        
    # Formulae to calculate average change per month by summing all the values in the diff_list and dividing them by count. Use Max and Min to store the max and min change.
    Sum = sum(diff_list)    
    average = round(Sum/count,2) 
    Max = max(diff_list)          
    Min = min(diff_list)
       
    # To identify the month (from column 1 of the csv) for max and min change, use the index function to find out the row number where they are located. Note this row number does not include the header, so add 1 to those numbers.
    Max_month = diff_list.index(Max)
    Min_month = diff_list.index(Min)

    # Print the calculated values in the format required for the assignment. Greatest increase and decrease in profits are identify by using the max and min months + 1 as the index of the list of rows created as data_cols.
    print(" Financial Analysis\n -----------------------------")
    print(f" Total Months: {skip}")
    print(f" Total: ${Total}") 
    print(f" Average Change: ${average}")  
    print(f" Greatest Increase in Profits: {data_cols[Max_month+1]} $({Max})")     
    print(f" Greatest Decrease in Profits: {data_cols[Min_month+1]} $({Min})") 

# Specify the path for creating the output_report as a text file
outfile = Path('output_report.txt')

# Open the output file as ofile and write each of the above print commands sequentially. Remember to move to next line after each line using \n/ 
with open(outfile, 'w') as ofile:
    ofile.write(" Financial Analysis\n -----------------------------\n")
    ofile.write(f" Total Months: {skip}\n")
    ofile.write(f" Total: ${Total}\n")
    ofile.write(f" Average Change: ${average}\n")
    ofile.write(f" Greatest Increase in Profits: {data_cols[Max_month+1]} $({Max})\n")
    ofile.write(f" Greatest Decrease in Profits: {data_cols[Min_month+1]} $({Min})\n")   
    
```
