```python
## This is a script to calculate average, maximum and minimum of monthly changes in profits/loss from bugdet_data.csv

# Import libraries required to run the script
import csv
from pathlib import Path


skip = 0
count = 0
prev_row = 0
difference = 0
diff_list = []
average = 0
Max = 0
Min = 0
Total = 0
data_cols = []

data = Path('../../Resources/Instructions/PyBank/Resources/budget_data.csv')
# 
with open(data, 'r') as xfile:
    datafiles = csv.reader(xfile, delimiter=',')
    # Skip the first row
    header = next(datafiles)

    for row in datafiles:
        skip +=1
        Total += int(row[1])
        data_cols.append(row[0])
        if skip ==1:
            prev_row = int(row[1])            
        if skip >1: 
            count +=1            
            difference = int(row[1]) - prev_row
            diff_list.append(difference)
            prev_row = int(row[1])
        

    Sum = sum(diff_list)    
    average = round(Sum/count,2) 
    Max = max(diff_list)          
    Min = min(diff_list)
       
    

    Max_month = diff_list.index(Max)
    Min_month = diff_list.index(Min)


    print(" Financial Analysis\n -----------------------------")
    print(f" Total Months: {skip}")
    print(f" Total: ${Total}") 
    print(f" Average Change: ${average}")  
    print(f" Greatest Increase in Profits: {data_cols[Max_month+1]} $({Max})")     
    print(f" Greatest Decrease in Profits: {data_cols[Min_month+1]} $({Min})") 

outfile = Path('output_report.txt')

with open(outfile, 'w') as ofile:
    # outfile = csv.writer(ofile, delimiter=',')
    ofile.write(" Financial Analysis\n -----------------------------\n")
    ofile.write(f" Total Months: {skip}\n")
    ofile.write(f" Total: ${Total}\n")
    ofile.write(f" Average Change: ${average}\n")
    ofile.write(f" Greatest Increase in Profits: {data_cols[Max_month+1]} $({Max})\n")
    ofile.write(f" Greatest Decrease in Profits: {data_cols[Min_month+1]} $({Min})\n")   
    
```
